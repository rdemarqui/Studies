## Confidence Interval

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Definir os parâmetros
np.random.seed(42)  # Para reprodutibilidade
true_mean = 5  # Média verdadeira da população
std_dev = 2  # Desvio padrão da população
n_samples = 30  # Tamanho da amostra em cada experimento
n_experiments = 100  # Número de experimentos

# Armazenar os resultados
intervals = []
contains_true_mean = []

# Executar 100 experimentos
for _ in range(n_experiments):
    sample = np.random.normal(loc=true_mean, scale=std_dev, size=n_samples)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1) / np.sqrt(n_samples)  # Erro padrão da média

    # Intervalo de confiança de 95%
    ci_low, ci_high = stats.norm.interval(0.95, loc=sample_mean, scale=sample_std)
    intervals.append((ci_low, ci_high))

    # Verificar se o intervalo contém a média verdadeira
    contains_true_mean.append(ci_low <= true_mean <= ci_high)

# Visualização
plt.figure(figsize=(10, 6))
for i, (ci_low, ci_high) in enumerate(intervals):
    color = 'green' if contains_true_mean[i] else 'red'
    plt.plot([i, i], [ci_low, ci_high], color=color, marker='o')

plt.axhline(y=true_mean, color='blue', linestyle='--', label="Média verdadeira")
plt.xlabel("Experimento")
plt.ylabel("Intervalo de confiança")
plt.title("Intervalos de Confiança de 95% em 100 Experimentos")
plt.legend()
plt.show()

# Proporção de intervalos que contêm a média verdadeira
np.mean(contains_true_mean)
```

The chart above shows the 95% confidence intervals calculated from 100 independent experiments. The dashed blue line indicates the true population mean (μ = 5). The intervals colored green are those that contain the true mean, while the intervals colored red do not. In this example, approximately 93% of the confidence intervals contain the true mean, which is close to what is expected for a 95% confidence interval. This demonstrates that the frequentist definition is in line with the idea that, in repeated experiments, approximately 95% of the intervals would contain the true value.
