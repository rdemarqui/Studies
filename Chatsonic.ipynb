```python
import requests
```

```python
api_key = ""
```

```python
role = """
Utilizando somente as opções: 
Cobrança indevida, migração de plano, consumo de franquia, pagamento/recarga, atendimento, serviço indevido, 
internet lenta, rede instavel, sem sinal, portabilidade, ativação chip/sim, ligações telefonicas, SMS, sem classificação.

Classifique sem justificar, o texto com no máximo duas opções:
"""

complain = """
No dia 09-03-2023 entrei em contato com a Tim *144 porque estava com uma fatura no valor aproximado de 200,00 atrasada, pois nessa fatura incluía um parcelamento de pendências anteriores. Pois como não consegui pagar, pedi o reparcelamento e a moça que me atendeu disse que faria o reparcelamento incluído tudo. Então tive que dar a entrada de 112,00 com vencimento dia 13-03, mas paguei dia 09-03.
Porém recebi mensagens de cobrança e descobri que a fatura de 200 em atraso não estava incluído no acordo atual. Como ela não incluiu? Sendo que nessa fatura era o mesmo acordo que ela estava reparcelando, isto é, paguei a Tim achando que estava tudo ok, e continuei com dívida. Agora não tenho valor para dar outra entrada. Peço por gentileza que corrijam o problema
"""

input_text = role + '\n' + complain

if len(input_text) > 2000:
  input_text = input_text[:2000]
```

```python
print(input_text)
```

```python
url = "https://api.writesonic.com/v2/business/content/chatsonic?engine=premium"

payload = {
    "input_text": input_text,
    "enable_google_results": "true",
    "enable_memory": False
}
headers = {
    "accept": "application/json",
    "content-type": "application/json",
    "X-API-KEY": api_key
}

response = requests.post(url, json=payload, headers=headers)
```

```python
print(response.text)
```
